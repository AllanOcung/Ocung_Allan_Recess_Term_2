**Question 1:** write a function `isValid(s)` that takes as argument a string s containing a sequence of parenthesis '(', ')', '{', '}', '[' and ']', and  determines if the input is valid. A input string is valid if for every open parenthensis there is a close one and parenthesis is well-formed. e.g  "(){}[]" is valid.

```python
def isValid(s: str) -> bool:

    stack = []     # Stack to keep track of opening parantheses

    valid_Pairs = {')': '(', '}': '{', ']': '['} # Dict to hold matching pairs
    
    # Iterate over each char in the string
    for char in s:
        if char in valid_Pairs.values():

            # # If the character is an opening parenthesis, push it to the stack
            stack.append(char)

        elif char in valid_Pairs.keys():

            # If the character is a closing parenthesis
            if stack == [] or stack.pop() != valid_Pairs[char]:
                return False
        # Invalid char  
        else:
            return False 
        
    # Check if the stack is empty, it returns true when empty
    return stack == []

print(isValid('(]'))

print(isValid('(){}[]'))

print(isValid('[{}]'))

print(isValid('({[)]}'))

print(isValid('((()'))

print(isValid(''))
    
```

**Question 2:** Given a paragraph as a string, write a function that return the number of character with odd frequencies. E.g The paragraph ``DSA 2024 Nyeri`` has *10* characters with odd frequencies. i.e the entire frequency count is given as {' ': 2, '2': 2, 'D': 1, 'S': 1, 'A': 1, '0': 1, '4': 1, 'N': 1, 'y': 1, 'e': 1, 'r': 1, 'i': 1}) and there are *10* characters with odd frequences. So the function should return *10*.

```python
# Frequency count is the number of occurrences of a given char in a string

from collections import Counter

def oddFrequencyCounter(theParagraph: str) -> int:

    frequency = Counter(theParagraph) # Count the freq of each char

    # count the number of chars with odd freqs
    odd_freq_count = sum(1 for count in frequency.values() if count % 2 != 0)

    return odd_freq_count

paragraph = 'Frequency count is the number of occurrences of a given char in a string'

print(f'Number of characters with odd frequencies: {oddFrequencyCounter(paragraph)}')

```

**Question 6:**  Write a function `greatest_common_divisor` that takes two inputs `a` and `b` and returns the greatest common divisor of the two numbers. E.g. input `(10, 15)` would return `5`

```python
# To compute the greatest common divisor (GCD) of two numbers, you can use the Euclidean algorithm.
'''
In mathematics, the Euclidean algorithm,[note 1] or Euclid's algorithm, is an efficient method 
for computing the greatest common divisor (GCD) of two integers (numbers), the largest number 
that divides them both without a remainder.
'''

def greatest_common_divisor(a: int, b: int) -> int:
    while b:

        # Update a to b and b to a % b
        a, b = b, a % b

    # When b is 0, a contains the GCD
    return a

print(greatest_common_divisor(10, 15))

print(greatest_common_divisor(7, 3))

print(greatest_common_divisor(100, 85))

print(greatest_common_divisor(21, 14))
```

**Question 7:**  Write a function `get_3_nearest` that takes in a point of interest ``pt`` and a **list** of points ``ptlist``  and returns a list of 3 nearest points from the point of interest ``pt``. Assume the distance between any two point is defined by the `L1-norm`.

```python
def manhattan_distance(p1, p2):

    # Calculate distances from pt to each point in ptlist
    return abs(p1[0] - p2[0] + abs(p1[1] - p2[1]))


def get_3_nearest(pt, ptlist):

    distances = [(point, manhattan_distance(pt, point)) for point in ptlist]

    # sort points based on distances
    distances.sort(key=lambda x: x[1])

    # return the 3 nearest points
    return [distances[i][0] for i in range(min(3, len(distances)))]

pt = (0, 0)

ptlist = [(1, 1), (2, 2), (-1, 1), (0, 2), (3, 0)]

nearest_points = get_3_nearest(pt, ptlist)

print(f'3 Nearest Points to {pt}, are: {nearest_points}')
    
```

**Question 8:**  Write a function `diagonal_vector(M)` that returns a **numpy** array of the list of **absolute** values of the main diagonal entries in the matrix $M$

```python
def diagonal_vector(M):

    # Extract the main diagonal entries
    main_diagnol = np.diagonal(M)

    # Compute the absolute values of the main diagonal entries
    abs_diagnol = np.abs(main_diagnol)

    # Return the result as a numpy array
    return abs_diagnol 

M = np.array([[3, -2, 3], [4, -5, 6], [-7, 8, 9]])

result = diagonal_vector(M)

print(result)
```

**Question 9:**  Write a function `flatten_reverse_lists` that takes in a list of lists and outputs a **reverse** sorted list of elements of sublists of the input list (confusing right?) <br>
Example: given `flatten_reverse_lists([[2,13,44], [6,7]])` it should return `[2,6,7,13,44]`

```python
def flatten_reverse_lists(superlist):

    # Flatten the list of lists into a single list
    flat_list = [item for sublist in superlist for item in sublist]

    sorted_list = sorted(flat_list)

    return sorted_list

input_list = [[2,13, 44], [6, 7]]

result = flatten_reverse_lists(input_list)

print(result)
```

**Question 10:** Do the following to the dataframe:
* Create a new collumn ``total_price`` whose value is equal to $scoops * price$*
* Write a function ``groupStatistics(data, groupValue)``. Internally, this function groups ``data``  by ``flavor`` and then returns statistics of a given grouped item ``groupValue`` indexed on the ``total_price`` columns. The statistics is a numpy array contains ``[mean, media, min, max, std]`` of the ``total_price`` column. The ``std`` should be rounded to 2 **decimal places**



```python
data['total_price'] = data['scoops'] * data['price']

print("Enhanced DataFrame:\n")

print(data)

def groupStatistics(data, groupValue):
    group = data.groupby('flavor')['total_price'].agg(['mean', 'median', 'min', 'max', 'std'])

    group['std'] = round(group['std'], 2)

    return group.loc[groupValue].values


flavor = 'dark chocolate'

stats = groupStatistics(data, flavor)

print(f"\nStatistics for '{flavor}':\n")
print(f"Mean: {stats[0]}\n")
print(f"Median: {stats[1]}\n")
print(f"Min: {stats[2]}\n")
print(f"Max: {stats[3]}\n")
print(f"Standard Deviation: {stats[4]}")
```
