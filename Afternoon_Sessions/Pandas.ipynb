```python
# Pandas series

# What is a Series?

# A Pandas Series is like a column in a table.

# It is a one-dimensional array holding data of any type.

# Creating a simple Pandas Series from a list

import pandas as pd

z = [1, 7, 2]

myVar = pd.Series(z)

print(myVar)

# Return the first value of the Series

print(myVar[1])
```

```python
import pandas as pd

# Create Labels
# With the index argument, you can name your own labels.

z = [1, 7, 2]

myVar = pd.Series(z, index = ["x", "y", "z"])

print(myVar)

# Return the value of z

print(myVar['z'])

```

```python
import pandas as pd

# Create a simple Pandas Series from a dictionary
# The keys of the dictionary become the labels

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)

# Selecting only some of the items in the dict, 
# use the index argument and specify only the items you want to include in the Series.

myvar = pd.Series(calories, index = ['day1', 'day3'])

print(myvar)
```

```python
# DataFrames

# Data sets in Pandas are usually multi-dimensional tables, called DataFrames.

# Series is like a column, a DataFrame is the whole table.

# A Pandas DataFrame is a 2 dimensional data structure

# Creating a DataFrame from two Serires

import pandas as pd

data = {
    'calories': [420, 380, 390],
    'duration': [50, 40, 45]
}

# load data into a DataFrame object
myvar = pd.DataFrame(data, index = ['day1', 'day2', 'day3'])

print(myvar)

print()
# Locate Row

# Return row 1. Returns a Pandas Series
print(myvar.loc['day2'])

print()

# Return row 0 and 1. Using a list of indexes. 
# When using [], the result is a Pandas DataFrame
print(myvar.loc[['day1', 'day3']])
```

```python
# Read CSV Files
# A simple way to store big data sets is to use CSV files (comma separated files).

import pandas as pd

df = pd.read_csv('data.csv')

# use to_string() to print the entire DataFrame.
print(df.to_string())
```
