```python
# Pyplot
# Most of the Matplotlib utilities lies under the pyplot submodule, 
# and are usually imported under the plt alias:


# Matplotlib Plotting
import matplotlib.pyplot as plt
import numpy as np

import otter

grader = otter.Notebook()


# Draw a line in a diagram from position (0,0) to position (6,250):
xpoints = np.array([1, 8])

ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)

plt.show()



```

```python
# Draw two points in the diagram, one at position (1, 3) and one in position (8, 10):

xpoints = np.array([1, 8])

ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')

plt.show()

```

```python
# Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

xpoints = np.array([1, 2, 6, 8])

ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)

plt.show()
```

```python
# Default X-Points
# If we do not specify the points on the x-axis, they will get the 
# default values 0, 1, 2, 3 etc., depending on the length of the y-points.

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)

plt.show()
```

```python
# Markers
# You can use the keyword argument marker to emphasize 
# each point with a specified marker:

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms=20)

plt.show()
```

```python
# Matplotlib Line

# Linestyle
# You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:

ypoints = np.array([3, 8, 1, 10])

# plt.plot(ypoints, linestyle='dotted') # 'dashed'

plt.plot(ypoints, ls=':') # 'dashed = --'

plt.show()
```

```python
# Line Color
# You can use the keyword argument color or the shorter c to set the color of the line:

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color='r', marker='o')


plt.show()
```

```python
# Multiple Lines

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1, color='b', marker='o')
plt.plot(y2, color='r', marker='o')


plt.show()
```

```python
# Or 

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
# Matplotlib Labels and Title

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title('Sports Watch Data')

plt.xlabel('Average Pulse')

plt.ylabel('Calorie Burnage')

plt.show()
```

```python
# Set font properties for the title and labels

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])

y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family': 'serif', 'color':'blue', 'size':20}
font2 = {'family': 'serif', 'color':'darkred', 'size':15}

plt.title('Sports Watch Data', fontdict=font1)

plt.xlabel('Average Pulse', fontdict=font2)

plt.ylabel('Calorie Burnage', fontdict=font2)

plt.plot(x, y)

# plt.grid() # ploting grid lines, you can also specify which grid lines to show: plt.grid(axis='x')

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

```python
# Matplotlib Subplot

# Plot 1:

x = np.array([0, 1, 2, 3])

y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)

plt.plot(x, y)

plt.title("SALES")

# Plot 2:

x = np.array([0, 1, 2, 3])

y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)

plt.plot(x, y)

plt.title("INCOME")

plt.suptitle("MY SHOP") # Super Title
plt.show()
```
