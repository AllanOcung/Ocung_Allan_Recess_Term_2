```python
# Step 1: import neccessary libraries

import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
# step 2: Load dataset

df = pd.read_csv('Ocung_Allan_datasets/niceguys.csv')

print(df.to_string())
```

```python
# Step 3: Data cleaning and preprocessing

print('Summary of the data:\n')
print(df.info())

print()

```

```python
df.describe(include='all')
```

```python

# Check for missing values
print('Number of null values per column:\n')

print(df.isnull().sum())
```

```python
# Handling misssing values

# Fill numerical columns with mean or median
df['age'].fillna(df['age'].median(), inplace=True)

df['living_allowance'].fillna(df['living_allowance'].mean(), inplace=True)


# Filling categorical columns with mode
df['number_of_siblings'].fillna(df['number_of_siblings'].mode()[0], inplace=True)

df['meals_per_day'].fillna(df['meals_per_day'].mode()[0], inplace=True)

df['has_boyfriend_girlfriend'].fillna(df['has_boyfriend_girlfriend'].mode()[0], inplace=True)

df['part-time_job'].fillna(df['part-time_job'].mode()[0], inplace=True)

print(df.isnull().sum())

```

```python
# Fix Incorrect Formats:
# Convert non-numeric strings to numbers:

# Define a mapping for common string representations of numbers
num_mapping = {'one': 1, 'two': 2, 'three': 3, 'four': 4, 'ninety-seven': 97}

# Apply the mapping to relevant columns
df['number_of_siblings'] = df['number_of_siblings'].replace(num_mapping).astype(int)

df['meals_per_day'] = df['meals_per_day'].replace(num_mapping).astype(int)

df['attendance_rate'] = df['attendance_rate'].replace(num_mapping).astype(float)

print(df.to_string())

```

```python
df = df.dropna()

df = df.reset_index(drop=True)

print(df.to_string())
```

```python
# Validate DeprecationWarning
print(df['sex'].unique())
print(df['has_boyfriend_girlfriend'].unique())
print(df['part-time_job'].unique())


# Ensure numerical columns have realistic values
print(df.describe().to_string())
```

```python
# Handling inconsistent data

# Convert to uppercase to avoid 'F' vs 'f' issues
df['sex'] = df['sex'].str.upper()  

df['has_boyfriend_girlfriend'] = df['has_boyfriend_girlfriend'].str.capitalize()

df['part-time_job'] = df['part-time_job'].str.capitalize()


# Save the cleaned data
df.to_csv('cleaned_niceguys.csv', index=False)
```

```python
# Exploratory data analysis (EDA)

# Exploratory data analysis (EDA) is used by data scientists to analyze 
# and investigate data sets and summarize their main characteristics, 
# often employing data visualization methods.
```

```python
# 1: Load Necessary Libraries and Dataset

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as pl
import seaborn as sns
import numpy as np

d = pd.read_csv('Ocung_Allan_datasets/cleaned_niceguys.csv') # Load the cleaned dataset
```

```python
# 2: Initial Exploration

print(d.to_string())
```

```python
# Summary statistics

print(d.describe().to_string())
```

```python
# Data types and missing values

print(d.info())
```

```python
# 3: Univariate Analysis

# a.Numerical Variables

# Histograms

plt.figure(figsize=(10, 6))
sns.histplot(d['age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplots

plt.figure(figsize=(8, 5))
sns.boxplot(x='course', y='living_allowance', data=d)
plt.title('Boxplot of Living Allowance by Course')
plt.xticks(rotation=45)
plt.show()
```

```python
# b.Categorical Variables

# Countplots

plt.figure(figsize=(8, 5))
sns.countplot(x='sex', data=df, hue='sex', palette='Set2', dodge=False, legend=False)
plt.title('Count of Students by Sex')
plt.xlabel('sex')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Bivariate and Multivariate Analysis

# Scatterplots

plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='living_allowance', data=d, hue='has_boyfriend_girlfriend', palette='viridis')
plt.title('Scatterplot of Age vs Living Allowance')
plt.xlabel('Age')
plt.ylabel('Living Allowance')
plt.legend(title='Has Boyfriend/Girlfriend')
plt.savefig('Scatterplot.png')
plt.show()
```

```python
# Pairplot (for exploring relationships between multiple variables)

sns.pairplot(d[['age', 'living_allowance', 'number_of_siblings', 'meals_per_day', 'grade']], diag_kind='kde')
plt.suptitle('Pairplot of Numerical Variables')
plt.savefig('Pairplot.png')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(d[['age', 'living_allowance', 'number_of_siblings', 'meals_per_day']].corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap_plot.png')
plt.show()
```
